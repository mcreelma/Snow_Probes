# Troubleshoot the ',' problem

```python
import numpy as np
dfnew = pd.DataFrame()
dfnew['Description'] = df2['Description']

dfnew['Depth'] = np.nan
dfnew['Mass'] = np.nan

for i in range(len(dfnew['Description'])):

  # if the object is an interger
  if type(dfnew['Description'][i]) == int:
    # Enter it as a depth
    dfnew['Depth'][i] = dfnew['Description'][i]
    dfnew['Description'][i] = np.nan
    # print('isnan')

  # If the object is a string: 
  elif type(dfnew['Description'][i]) != int:

    # isolate that into a single variable
    string = dfnew['Description'][i] 
    
    try: # try to split the string by the comma
      [v1 , v2] = string.split(',') # separate the two by the comma
      dfnew['Depth'][i] = int(v1) # enter v1 as a depth
      # enter the first 4 characters of the string as a density 
      dfnew['Mass'][i] = int(v2[0:4]) 
      dfnew['Description'][i] = v2[-1:5]

    except: # if the string can't be split
      dfnew['Depth'][i] = int(string) # enter v1 as a depth
      dfnew['Mass'][i] = 'NaN'
      dfnew['Description'][i] = np.nan

# alter the df2 with the new data
df2['Depth'] = dfnew['Depth']
df2['Mass'] = dfnew['Mass']
df2.sample(10)


# append to new dataframe 
df['Mass'] = np.nan
df = df.append(df2, ignore_index = True)
```

# Create geopandas dataframe

```python
# Convert the dataframe
gdf = gpd.GeoDataFrame(df, 
                       geometry=gpd.points_from_xy(df.Longitude, df.Latitude),
                       )

# set the order of the columns
gdf = gdf [['Source' , 'Name' , 'Depth' , 'Altitude (ft)' , 'Accuracy (ft)' , 'Date/Time',
           'Latitude' , 'Longitude' , 'Location' , 'geometry' , 'Description' , 'User' , 'Attachment'	]]

# set the crs
gdf = gdf.set_crs(4326, allow_override=True)
display(gdf)
print(gdf.crs)
```
